```python
import tushare as ts
import pandas as pd
import numpy as np
```

```python
api='7f3dda5fe2c762a77fdc4ff58496205e0aac3ced9eab0f6333b80dcd'#调用api获取股票数据
type_stock='000063.SZ'#股票代码
start_date='20230101'#开始日期
end_date='20231231'#结束日期
pro = ts.pro_api(api)
stock= pro.daily(ts_code=type_stock,start_date=start_date,end_date=end_date)
stock['trade_date']=pd.to_datetime(stock['trade_date'])
df1=stock.set_index('trade_date')
df1=df1.sort_index()
df1.rename(columns={'vol':'volume'},inplace=True)
df1
```
